```python
genres_list = df3.genres.to_list()

genre_count = {}
for genres in genres_list:
  try:
    genre_list = eval(genres)
    for genre in genre_list:
      if genre in genre_count.keys():
        genre_count[genre] += 1
      else:
        genre_count[genre] = 1
  except:
    pass
  

print(f"Number of Genres: {len(genre_count)}")
print(genre_count)
```

```python
df3 = df3.dropna().reset_index(drop=True)
```

```python
df3 = df3.drop_duplicates(subset=["title"], keep=False)
print(len(df3))
```
